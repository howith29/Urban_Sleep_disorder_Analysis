```python
import pandas as pd
```

```python
col = ['신고년도','시도','시군구','총신고수','가정폭력신고수','데이트폭력신고수','성폭력신고수','몰래카메라신고수','살인신고수','강도신고수','절도신고수','폭력신고수']
```

```python
df1 = pd.read_csv('data2/범죄 빈발지 데이터 2019.csv', names=col, header=None)
df3 = pd.read_csv('data2/범죄 빈발지 데이터 2021.csv', names=col, header=None)
df4 = pd.read_csv('data2/범죄 빈발지 데이터 2022.csv', names=col, header=None)
df5 = pd.read_csv('data2/범죄 빈발지 데이터 2023.csv', names=col, header=None)
```

```python
df1
```

```python
# 서울, 경기도, 인천광역시 데이터만 추출
valid_regions = ['서울특별시', '경기도', '인천광역시']
df1 = df1[df1['시도'].isin(valid_regions)]
df3 = df3[df3['시도'].isin(valid_regions)]
df4 = df4[df4['시도'].isin(valid_regions)]
df5 = df5[df5['시도'].isin(valid_regions)]
```

```python
def remove_junggu(df):
    df.loc[(df['시도']=='서울특별시')&(df['시군구']=='중구'), '시군구'] = '서울중구'
    df.loc[(df['시도']=='인천광역시')&(df['시군구']=='중구'), '시군구'] = '인천중구'
    return df
```

```python
df1 = remove_junggu(df1)
df3 = remove_junggu(df3)
df4 = remove_junggu(df4)
df5 = remove_junggu(df5)
```

```python
df1
```

```python
def col_remove(df):
    df.drop(columns=['신고년도','가정폭력신고수','데이트폭력신고수','성폭력신고수','몰래카메라신고수','살인신고수','강도신고수','절도신고수','폭력신고수'], inplace=True)
    return df
```

```python
df1 = col_remove(df1)
df3 = col_remove(df3)
df4 = col_remove(df4)
df5 = col_remove(df5)
```

```python
make_code = lambda x: x + '1'
df1['시군구코드'] = df1['시군구'].apply(make_code)
df5['시군구코드'] = df5['시군구'].apply(make_code)
df3['시군구코드'] = df3['시군구'].apply(make_code)
df4['시군구코드'] = df4['시군구'].apply(make_code)

```

```python
# 데이터 병합
crime = pd.concat([df1, df3, df4, df5], axis=0)
```
