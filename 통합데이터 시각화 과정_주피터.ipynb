코사인 유사도 측정

```python
#컬럼 리스트화
patient_list = t_data['진료인원(지수)'].tolist()
patient_list_seoul = data_seoul['진료인원(지수)'].tolist()
order_list = t_data['총 주문건수(지수)'].tolist()
alcohol_list = t_data['유흥업소(지수)'].tolist()
act_list = t_data['인당시설수(지수)'].tolist()
crime_list = t_data['범죄율(지수)'].tolist()
civil_list = data_seoul['도심화(지수)'].tolist()
estate_list = t_data['공시지가(지수)'].tolist()


from sklearn.metrics.pairwise import cosine_similarity

# 코사인 유사도 계산
similarity_1_2 = cosine_similarity([patient_list], [order_list])[0][0]  # 환자수 / 배달
similarity_1_3 = cosine_similarity([patient_list], [alcohol_list])[0][0]  # 환자수 / 유흥업소
similarity_1_4 = cosine_similarity([patient_list], [act_list])[0][0]  # 환자수 / 활동량
similarity_1_5 = cosine_similarity([patient_list], [crime_list])[0][0]  # 환자수 / 범죄율
similarity_1_6 = cosine_similarity([patient_list_seoul], [civil_list])[0][0]  # 환자수 / 도심화
similarity_1_7 = cosine_similarity([patient_list], [estate_list])[0][0]  # 환자수 / 공시지가


# 결과 출력
print(f"환자수 & 배달주문량 코사인 유사도: {similarity_1_2:.4f}")
print(f"환자수 vs 유흥업소 코사인 유사도: {similarity_1_3:.4f}")
print(f"환자수 vs 활동량 코사인 유사도: {similarity_1_4:.4f}")
print(f"환자수 vs 범죄율 코사인 유사도: {similarity_1_5:.4f}")
print(f"환자수(서울) vs 도심화 코사인 유사도: {similarity_1_6:.4f}")
print(f"환자수 vs 공시지가 코사인 유사도: {similarity_1_7:.4f}")
```
