결측치 처리 과정

```python
#결측치 처리

#유흥업소 결측치는 위 데이터값으로 채움
t_data['유흥업소 개수'] = t_data['유흥업소 개수'].bfill()

#평균주문금액 결측치는 평균값으로 채움
mean1 = t_data['평균주문금액'].mean()
t_data['평균주문금액'] = t_data['평균주문금액'].fillna(mean1)

#유입야간교통량평균 결측치는 평균값으로 채움
mean2 = t_data['유입야간교통량평균'].mean()
t_data['유입야간교통량평균'] = t_data['유입야간교통량평균'].fillna(mean2)
```

```python
# 총 주문건수 컬럼 생성

t_data['총 주문건수'] = t_data[['1세대 가구' ,'1인 가구', '2세대 가구' ,'3세대 가구' ,'4세대 이상 가구' ,'비친족 가구']].sum(axis=1)

t_data
```

```python
t_data.columns
```

데이터 지수화 작업

```python
# 데이터 지수화 작업 (그래프 코사인 유사도 측정을 위해 함수값 범위 조정)
# 컬럼별 최대값 산출

max_h = t_data['진료인원(명)'].max()
max_d = t_data['총 주문건수'].max()
max_a = t_data['유흥업소 개수'].max()
max_f = t_data['시군구별시설수'].max()
max_p = t_data['시군구별인구수'].max()
max_e = t_data['인당시설수'].max()
max_c = t_data['총신고수'].max()
max_t = t_data['유입야간교통량평균'].max()


t_data['진료인원(지수)'] = t_data['진료인원(명)'].apply(lambda x : x / max_h)

t_data['총 주문건수(지수)'] = t_data['총 주문건수'].apply(lambda x : x / max_d)

t_data['유흥업소(지수)'] = t_data['유흥업소 개수'].apply(lambda x : x / max_a)

t_data['시설수(지수)'] = t_data['시군구별시설수'].apply(lambda x : x / max_f)

t_data['인구수(지수)'] = t_data['시군구별인구수'].apply(lambda x : x / max_p)

t_data['인당시설수(지수)'] = t_data['인당시설수'].apply(lambda x : x / max_e)

t_data['범죄율(지수)'] = t_data['총신고수'].apply(lambda x : x / max_c)

t_data['도심화(지수)'] = t_data['유입야간교통량평균'].apply(lambda x : x / max_t)
```

```python
t_data = t_data.reindex(['시도', '시군구2', '주상병명', '진료인원(명)', '진료인원(지수)','1세대 가구', '1인 가구', '2세대 가구', '3세대 가구',
       '4세대 이상 가구', '비친족 가구','총 주문건수(지수)','평균주문금액','유흥업소 개수','유흥업소(지수)','시군구별시설수','시설수(지수)',
       '시군구별인구수','인구수(지수)','인당시설수','인당시설수(지수)','총신고수','범죄율(지수)','유입야간교통량평균','도심화(지수)',
       '공시지가', '공시지가2'
       ], axis=1)
```
