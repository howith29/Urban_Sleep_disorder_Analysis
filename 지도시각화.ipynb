```python
import geopandas as gpd
import json
import folium
```

```python
# GeoJSON 파일 불러오기 (서울 + 인천 + 경기기)
with open(r'C:\Users\user\Documents\데이터 분석 기획 수업\수업자료\데이터 분석 수업\1차 프로젝트\1차 프로젝트\지도 히트맵 관련\geo json\hangjeongdong_서울특별시.geojson', encoding="utf-8") as f:
    seoul_geo = json.load(f)

with open(r'C:\Users\user\Documents\데이터 분석 기획 수업\수업자료\데이터 분석 수업\1차 프로젝트\1차 프로젝트\지도 히트맵 관련\geo json\hangjeongdong_인천광역시.geojson', encoding="utf-8") as f:
    incheon_geo = json.load(f)

with open(r'C:\Users\user\Documents\데이터 분석 기획 수업\수업자료\데이터 분석 수업\1차 프로젝트\1차 프로젝트\지도 히트맵 관련\geo json\hangjeongdong_경기도.geojson', encoding="utf-8") as f:
    gyeongi_geo = json.load(f)

# 3개 지역 GeoJSON 데이터를 병합 (리스트 합치기)
merged_geojson = {
    "type": "FeatureCollection",
    "features": seoul_geo["features"] + incheon_geo["features"] + gyeongi_geo["features"]
}

merged_geojson['features'][0]['properties']
```

```python
# GeoJSON 내 지역명 리스트 (중복 제거 후 정렬)
geo_names = sorted(set(feature["properties"]["sggnm"] for feature in merged_geojson["features"]))
print(" GeoJSON 지역명 리스트:", geo_names)

# 데이터프레임 내 지역명 리스트 (중복 제거 후 정렬)
t_data_names = sorted(set(t_data["시군구"].unique()))
print("\n 데이터프레임 시군구 리스트:", t_data_names)
```

```python
# 1. 구 단위를 시 단위로 변환하는 매핑 딕셔너리 생성
gu_to_si = {
    "고양시덕양구": "고양시", "고양시일산동구": "고양시", "고양시일산서구": "고양시",
    "수원시권선구": "수원시", "수원시영통구": "수원시", "수원시장안구": "수원시", "수원시팔달구": "수원시",
    "성남시분당구": "성남시", "성남시수정구": "성남시", "성남시중원구": "성남시",
    "안산시단원구": "안산시", "안산시상록구": "안산시",
    "용인시기흥구": "용인시", "용인시수지구": "용인시", "용인시처인구": "용인시",
    "안양시동안구": "안양시", "안양시만안구": "안양시"
}

# 2. 중구에 대한 매핑 추가
for feature in merged_geojson["features"]:
    old_name = feature["properties"]["sggnm"]  # 원래 지역명
    full_name = feature["properties"].get("adm_nm", "")  # 서울 or 인천 확인

    # 기존 매핑 적용
    if old_name in gu_to_si:
        feature["properties"]["sggnm"] = gu_to_si[old_name]

    # 서울 중구 처리
    elif old_name == "중구" and "서울특별시" in full_name:
        feature["properties"]["sggnm"] = "서울중구"

    # 인천 중구 처리
    elif old_name == "중구" and "인천광역시" in full_name:
        feature["properties"]["sggnm"] = "인천중구"

# 3. 변경된 GeoJSON 지역명 확인
geo_names = sorted(set(feature["properties"]["sggnm"] for feature in merged_geojson["features"]))
print("변환된 GeoJSON 지역명 리스트:", geo_names)
```

```python
import folium

# 지도 객체 생성 (기본 중심 좌표 설정)
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# Choropleth 레이어 추가
folium.Choropleth(
    geo_data=merged_geojson,  # GeoJSON 데이터
    name="배달주문량",  # 레이어 이름
    data=t_data,  # 데이터프레임
    columns=["시군구", "총 주문건수(지수)"],  # 데이터프레임 내 컬럼 (지역명, 색상 값)
    key_on="feature.properties.sggnm",  # GeoJSON 내 지역명을 찾는 key
    fill_color="Blues",  # 색상 스케일
    fill_opacity=0.7,  # 투명도
    line_opacity=0.2,  # 경계선 투명도
    legend_name="배달주문"  # 범례 이름
).add_to(m)

# 지도 컨트롤 추가
folium.LayerControl().add_to(m)

# 지도 출력
m.save("배달주문량.html")
print("지도 저장 완료: choropleth_map.html")
```
